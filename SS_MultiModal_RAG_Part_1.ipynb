```python
import os
import yt_dlp

def download_video_and_get_info(video_url, destination_path):
    # Ensure the target directory exists before running
    if not os.path.exists(destination_path):
        os.makedirs(destination_path)

    # Configure options: set the target download path and filename structure
    ydl_opts = {
        'outtmpl': os.path.join(destination_path, '%(title)s.%(ext)s'),
    }

    # Initialize YoutubeDL with configuration options
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        # Step 1: Extract video metadata
        # download=True tells it to immediately download the file after extracting info
        info_dict = ydl.extract_info(video_url, download=True)

        # Step 2: Safely read required information fields from the dictionary
        video_title = info_dict.get('title', 'Unknown Title')
        video_author = info_dict.get('uploader', 'Unknown Author')  # yt-dlp maps creator/channel to 'uploader'
        video_views = info_dict.get('view_count', 0)
        metadata = {"Author": video_author, "Title": video_title, "Views": video_views}

        # Print the extracted data points
        print("\n=== Video Information ===")
        print(f"Title:  {video_title}")
        print(f"Author: {video_author}")
        print(f"Views:  {video_views:,}")
        print(f"Metadata: {metadata}")

    return metadata




```

```python
from moviepy.editor import VideoFileClip
def video_to_images(video_path, output_folder):
  clip = VideoFileClip(video_path)
  clip.write_images_sequence(
      os.path.join(output_folder,"frame%04d.png"), fps=0.2
  )
```

```python
def video_to_audio(video_path, output_audio_path):
  clip = VideoFileClip(video_path)
  audio=clip.audio
  audio.write_audiofile(output_audio_path)
```

```python
def audio_to_text(audio_path):
  recognizer=sr.Recognizer()
  audio=sr.AudioFile(audio_path)

  with audio as source:
    audio_data=recognizer.record(source)

    try:

      #recognize the speech
      text = recognizer.recognize_whisper(audio_data)

    except sr.UnknownValueError:
      print("Speech recognition could not understand the audio.")
  return text
```

```python
from transformers import AutoConfig
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

mm_embed_model = HuggingFaceEmbedding(
    model_name="BAAI/BGE-VL-base",
    trust_remote_code=True
)
```

```python
text_embed_model = HuggingFaceEmbedding(
    model_name="BAAI/bge-base-en-v1.5"
)
```

```python
index = MultiModalVectorStoreIndex.from_documents(documents, storage_context=storage_context, embed_model=text_embed_model, image_embed_model=mm_embed_model)
```

```python
retriever_engine = index.as_retriever(similarity_top_k=1, image_similarity_top_k=5)
```

```python
from llama_index.core.response.notebook_utils import display_source_node
from llama_index.core.schema import ImageNode
```

```python
def retrieve(retriever_engine, query_str):
    retrieval_results = retriever_engine.retrieve(query_str)

    retrieved_image = []
    retrieved_text = []
    for res_node in retrieval_results:
        if isinstance(res_node.node, ImageNode):
            retrieved_image.append(res_node.node.metadata["file_path"])
        else:
            display_source_node(res_node, source_length=200)
            retrieved_text.append(res_node.text)

    return retrieved_image, retrieved_text
```

```python
img, text = retrieve(retriever_engine, query)
```
